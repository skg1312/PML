1. MAE (Mean Absolute Error) Calculation
   
Question:

Given the actual house prices [200000, 250000, 300000, 400000] and predicted prices 
[210000, 245000, 290000, 390000], write a Python function to compute the MAE.


```python

y_true = np.array([200000, 250000, 300000, 400000])
y_pred = np.array([210000, 245000, 290000, 390000])

mae = mean_absolute_error(y_true,y_pred)
print(f'MAE :{mae:.4f}')
```

2. RMSE (Root Mean Squared Error) Calculation

Question:

For actual exam scores [90, 85, 70, 60, 75] and predicted scores [88, 82, 65, 67, 
78], compute the RMSE using Python.

```python
y_true = np.array([90, 85, 70, 60, 75])
y_pred = np.array( [88, 82, 65, 67, 78])

rmse = mean_squared_error(y_true,y_pred)**0.5

print(f'RMSE : {rmse:.4f}')
```

3. R² Score (Coefficient of Determination)

Question:

You are given:

Write code to compute the R² score.

```python
y_true = np.array([10, 12, 14, 16, 18])
y_pred = np.array([11, 13, 13, 15, 19])

r2 = r2_score(y_true,y_pred)

print(f'R² : {r2:.4f}')
```

6. Evaluate a Regression Tree with MAE, RMSE, R²

Question:

Generate regression data using make_regression(n_samples=50, n_features=1, 
noise=10)
Train a Decision Tree Regressor, make predictions, and evaluate the model using MAE, 
RMSE, and R² Score

```python
X, y =  make_regression(n_samples=50, n_features=1, noise=10)

X_train , X_test,y_train,y_test = train_test_split(X,y,test_size=0.4)

reg = DecisionTreeRegressor(random_state=0)

reg.fit(X_train,y_train)

y_pred = reg.predict(X_test)

print(f'MSE of DTR : {mean_absolute_error(y_test,y_pred):.2f}')
print(f'RMSE of DTR : {mean_squared_error(y_test,y_pred)**0.5:.2f}')
print(f'R² Score : {r2_score(y_test,y_pred):.2f}')

```

8. CART (Classification and Regression Tree) for Classification

Question:

Using the Iris dataset (from sklearn.datasets import load_iris), implement a CART 
classifier.
Split the data, train the model, and print the accuracy, precision, and F1-score

```python
from sklearn.datasets import load_iris
from sklearn.metrics import accuracy_score, precision_score, f1_score, classification_report
iris = load_iris()
X = iris.data
y = iris.target
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)

CART = DecisionTreeClassifier(random_state=0)

CART.fit(X_train,y_train)

y_pred = CART.predict(X_test)

print(f'Accuracy : {accuracy_score(y_test, y_pred):.2f}')
print(f'Precision : {precision_score(y_test,y_pred,average="weighted"):.2f}')
print(f'F1-Score : {f1_score(y_test,y_pred,average="weighted"):.2f}')
print(f"Classification Report:{classification_report(y_test, y_pred)}")
```

9. CART for Regression with Visualization

Question:

Use make_regression to create data.
Train a DecisionTreeRegressor, and plot the predicted vs actual values to visually assess the 
regression tree.

```python
from sklearn.tree import plot_tree

X, y =  make_regression(n_samples=50, n_features=1, noise=10)

X_train , X_test,y_train,y_test = train_test_split(X,y,test_size=0.4)

reg = DecisionTreeRegressor(random_state=0)

reg.fit(X_train,y_train)

y_pred = reg.predict(X_test)

plt.scatter(y_test, y_pred, color="blue", label="Predicted vs Actual")
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], color="red", label="Perfect Fit", linestyle="--")
plt.xlabel("Actual Values")
plt.ylabel("Predicted Values")
plt.title("Decision Tree Regressor: Predicted vs Actual Values")
plt.legend()
plt.grid()
plt.show()
```

10. Decision Boundary Visualization for Classification Tree

Question:

Create a classification dataset using make_classification(n_samples=200, 
n_features=2).
Train a Decision Tree Classifier and plot the decision boundary using matplotlib

```python

import matplotlib.pyplot as plt

X, y = make_classification(n_samples=200, n_features=2, n_classes=2, n_informative=2, n_redundant=0, random_state=42)

clf = DecisionTreeClassifier(random_state=42)
clf.fit(X, y)

x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.01), np.arange(y_min, y_max, 0.01))

Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

plt.contourf(xx, yy, Z, alpha=0.8, cmap=plt.cm.Paired)
plt.scatter(X[:, 0], X[:, 1], c=y, edgecolor="k", cmap=plt.cm.Paired)
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.title("Decision Tree Classifier Decision Boundary")
plt.show()
```
