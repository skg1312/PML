```python
from sklearn.datasets import load_iris
import pandas as pd

# Load the dataset
iris = load_iris()

# Convert to pandas DataFrame
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)

# Optionally add the target column
df['target'] = iris.target

# Get shape
rows, cols = df.shape

# Display
print(f"Number of rows: {rows}")
print(f"Number of columns: {cols}")
```

```python
from sklearn.datasets import fetch_openml
import pandas as pd

# Load the MNIST dataset (slow the first time, cached afterward)
mnist = fetch_openml('mnist_784', version=1)

# Convert to DataFrame
df = pd.DataFrame(mnist.data)
df['target'] = mnist.target  # Add the target column (digit labels)

# Check the shape
print("Shape of the DataFrame:", df.shape)

# Number of rows and columns
rows = df.shape[0]
cols = df.shape[1]

print(f"Number of rows (samples): {rows}")
print(f"Number of columns (features + target): {cols}")
```

```python
from sklearn.datasets import load_iris
import pandas as pd

# Load the iris dataset
iris = load_iris()

# Create a DataFrame with features
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)

# Add the target column (numerical class labels)
df['target'] = iris.target

# Optionally map numeric targets to species names
df['species'] = df['target'].map({0: 'setosa', 1: 'versicolor', 2: 'virginica'})

# Set options to display all rows and columns
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.width',None)

# Print the entire DataFrame
print(df)
```

```python
from sklearn.datasets import load_iris

data = load_iris()
print(data.feature_names)
print(data.target_names)
print(data.data[:5])
```

âœ… from sklearn.datasets import load_iris
This imports the function load_iris() from scikit-learn's built-in datasets module.

load_iris() is a utility function to load the Iris flower dataset, which is a famous dataset in machine learning.

âœ… data = load_iris()
This loads the Iris dataset into a variable named data.

data is a Bunch object (similar to a dictionary). It contains:

data: the features (flower measurements)

target: the labels (species)

feature_names: names of the features

target_names: names of the species

DESCR: description of the dataset

âœ… print(data.feature_names)
Prints the names of the input features (columns):


['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)', 'petal width (cm)']
These are the measurements of the iris flowers.

âœ… print(data.target_names)
Prints the target classes (types of iris flowers):


['setosa' 'versicolor' 'virginica']
These are the species the flowers belong to (classification labels).

âœ… print(data.data[:5])
Prints the first 5 rows of the feature data (each row is one flowerâ€™s measurements).

Output will look something like:


[[5.1 3.5 1.4 0.2]
 [4.9 3.0 1.4 0.2]
 [4.7 3.2 1.3 0.2]
 [4.6 3.1 1.5 0.2]
 [5.0 3.6 1.4 0.2]]
Each row has 4 values (sepal length, sepal width, petal length, petal width in cm).

ðŸ“ Summary:
Part	Meaning
feature_names	Names of input columns (features)
target_names	Names of the flower species
data	Actual feature values (as NumPy array)
target	Target labels (as integers: 0, 1, 2)


```python
# Step 1: Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
from google.colab import files

# Step 2: Upload and read the dataset
uploaded = files.upload()  # Upload Mall_Customers.csv
df = pd.read_csv("Mall_Customers.csv")

# Step 3: Select features for plotting (2D)
X = df[["Annual Income (k$)", "Spending Score (1-100)"]]

# Step 4: Scatter plot with row numbers
plt.figure(figsize=(10, 6))
plt.scatter(X["Annual Income (k$)"], X["Spending Score (1-100)"], color='lightblue')

# Add row numbers as labels
for i in range(len(X)):
    plt.text(X.iloc[i, 0], X.iloc[i, 1], str(i), color='black', fontsize=8, ha='center', va='center')

# Labels and title
plt.xlabel("Annual Income (k$)")
plt.ylabel("Spending Score (1-100)")
plt.title("Scatter Plot with Row Numbers")
plt.grid(True)
plt.show()
```
